```python
import matplotlib.pyplot as plt
import missingno as msno
import pandas as pd

# Cargar dataset
dino_df = pd.read_csv("../data/01_raw/prehistoric_dinosaur_profiles.csv")

# Informaci√≥n general
print(dino_df.shape)
print(dino_df.dtypes)
print(dino_df.head())

# Resumen valores faltantes
print(dino_df.isnull().sum())
print(dino_df.isnull().mean() * 100)

# Visualizaci√≥n de missing values con tama√±o controlado
msno.matrix(dino_df, figsize=(10,4))   # ancho 10, alto 4
plt.show()

msno.heatmap(dino_df, figsize=(8,4))  # ancho 8, alto 4
plt.show()

# Estad√≠sticas b√°sicas
print(dino_df.describe(include='all'))

# Comentarios y hallazgos
# Valores nulos leves en algunas columnas, tipos de datos correctos.
# Las distribuciones de clases se pueden analizar para identificar posibles sesgos.
```
